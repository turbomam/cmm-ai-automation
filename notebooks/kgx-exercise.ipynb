```python
from kgx.sink import TsvSink
from kgx.transformer import Transformer
```

```python
# Create the node dictionary
node = {
    "id": "CHEBI:15377",  # CURIE format ID
    "name": "water",
    "category": ["biolink:ChemicalEntity"],  # List of categories
}
```

```python
transformer = Transformer()
```

```python
# setup tsv writing

sink = TsvSink(
    owner=transformer,
    filename="kgx-exercise",
    format="tsv",
    node_properties={"id", "name", "category"},  # Which fields to include
)
```

```python
sink.write_node(node)
```

```python
sink.finalize()
```
